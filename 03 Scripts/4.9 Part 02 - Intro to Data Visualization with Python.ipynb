## 02. Histogram of 'order_hour_of_day'

```python
# Create a histogram of 'order_hour_of_day' column
hist = df_findings['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram shows a cluster of values in the area around 10 and 15, a very few amount of orders between 0-5, and a gradually decrease of orders starting around 17. From this we can infer that the popular hour for orders is late morning to early afternoon.

```python
# Export this chart as an image file
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_order_hod.png'))
```

## 04. Sampling Dataset Creation

```python
# Create a random sample by generating a seed
np.random.seed(4)
```

```python
# Create a list holding True/False values to test the np.randt.rand() <= .07
dev = np.random.rand(len(df_findings)) <= 0.7
```

```python
# Store 70% of the sample in the big dataframe
big = df_findings[dev]
```

```python
# Store 30% of the sample in the small dataframe
small = df_findings[~dev]
```

```python
# Obtaining the length of an object
len(df_findings)
```

```python
# Checking that the length of the big and small dataframes is equal to the overall length
len(big) + len(small)
```

This matches so I can continue.

```python
# Creating a smaller dataset of just the needed columns
df_findings_sample_set = small[['order_hour_of_day','prices']]
```

```python
# Check the sample dataframe
df_findings_sample_set.head()
```

## 05. Line Chart of 'prices' and 'order_hour_of_day'

```python
# Create a line chart using the data sample
line_chart = sns.lineplot(data = df_findings_sample_set, x = 'order_hour_of_day', y = 'prices')
```

```python
# Export this chart as an image file
line_chart.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_chart_prices_order_hod.png'))
```

## 07. Scatterplot of 'age' and 'income'

```python
# Creating a smaller dataset of just the needed columns
df_findings_sample_set_3= small[['age','income']]
```

```python
# Check the sample dataframe
df_findings_sample_set_3.head()
```

```python
# Creating scatterplot of 'age' and 'income'
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = df_findings_sample_set_3)
plt.title('Scatterplot of Age & Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

The scatterplot shows that from 20 to 40 age range that income density (the number of individuals earning within a specific income range) is relatively consistent. Most individuals have incomes ranging from $0 to $20,000. Around age 40, there's a noticeable increase in income density. The range of incomes increases, with more individuals earning between $0 and $30,000.\
\
This falls in line with a logical explanation that the younger individuals might represent entry-level positions, internships, or early career stages where salaries are generally lower with a significant increase in their earning potential as individuals reach mid-career at around 40.\
\
 Individuals in the 20 to 40 age range may have limited spending power due to their lower income levels. With the increase in income density, individuals around this age likely have higher spending power, enabling them to afford more or invest in higher-value goods and services.

```python
# Export this chart as an image file
scatterplot.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatterplot_age_income.png'))
```
