## 04. Define price_label Function & Apply to Dataframe & Get Max Value

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Printing the column names of the DataFrame to grab title of needed column
print(ords_prods_merge.columns)
```

```python
# In looking at the columns, I see that "order_hour_of_day" will help me achieve this task
# I start ay identifying the busiest hours of the day
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
data = {
    'order_hour_of_day': [10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19, 20, 7, 21, 22, 23, 6, 0, 1, 5, 2, 4, 3],
    'count': [2761760, 2736140, 2689136, 2662144, 2660954, 2618532, 2535202, 2454203, 2087654, 1718118, 1636502, 1258305, 976156, 891054, 795637, 634225, 402316, 290493, 218769, 115700, 87961, 69375, 53242, 51281]
}

df_ords_prods_merge = pd.DataFrame(data)

# Sorting the dataframe by the 'order_hour_of_day' column
df_ords_prods_merge_sorted = df_ords_prods_merge.sort_values(by='order_hour_of_day')

# Display the sorted data
print(df_ords_prods_merge_sorted)
```

In looking at the data this way we can that the highest volume of orders occurs during midday, particularly between 10 AM and 11 AM. The dataset shows the highest volume of orders actually occurring between 10 AM and 3 PM (not just 10 AM to 11 AM), with each hour during this period seeing counts over 2 million. Early morning hours from 12 AM (0) to 6 AM (6) show significantly fewer orders. These hours are the least active, with the number of orders gradually increasing as the morning progresses.

```python
# Checking the stats
stats = df_ords_prods_merge['order_hour_of_day'].describe()
print(stats)
```

### Determining the Threshhold
Fewest Orders: Hours less than the 25th percentile (5.75 rounded to 6).\
Average Orders: Hours between the 25th percentile (6) and 75th percentile (17.25 rounded to 17).\
Most Orders: Hours greater than the 75th percentile (17).

```python
# Create a function that uses these thresholds to assign labels based off of the determined threshhold

# Creating an empty list to temorarily hold the labels  
result_third = []


# Use a loop to assign appropriate labels based based on the value of the day
for hour in ords_prods_merge["order_hour_of_day"]: 
    if hour < 6:
        result_third.append('Fewest orders') 
    elif 6 <= hour <= 17:
        result_third.append('Average orders')
    else:
        result_third.append('Most orders')
```

```python
result_third
```

## 10. Create a new column containing these labels called “busiest_period_of_day.”

```python
# Add the results as a new column to the DataFrame
ords_prods_merge['busiest_period_of_day'] = result_third
```
